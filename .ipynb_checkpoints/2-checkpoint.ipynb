```python
## hill_climbing algorithm
graph={
    'A':['B','C'],
    'B':['A','D'],
    'C':['A','E'],
    'D':['B','E'],
    'E':['D','F'],
    'F':['E']
}
heuristic={
    'A':45,
    'B':30,
    'C':24,
    'D':25,
    'E':15,
    'F':0
}
start='A'
goal='F'
def hill_climb(start,goal):
    cur=start
    path=[cur]
    while cur!=goal:
        n=graph[cur]
        if not n:
            print("no neigbors present stuck at local maxima")
        next_node=min(n,key=lambda x:heuristic[x])
        if heuristic[next_node]>=heuristic[cur]:
            print("reached at local maxima")
            return path
        cur=next_node
        path.append(next_node)
    return path
result=hill_climb(start,goal)
if result and result[-1]==goal:
    print(f"shortest path from {start} to {goal} is :{result}")
else:
    print(f"no path found .reached at :{result[-1]}")
```

```python
## A* algorithm
graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 5), ('E', 12)],
    'C': [('F', 7)],
    'D': [('G', 3)],
    'E': [('G', 2)],
    'F': [('G', 5)],
    'G': []
}

# Heuristic values (h)
h = {
    'A': 10, 'B': 8, 'C': 5, 'D': 7,
    'E': 3, 'F': 3, 'G': 0
}
start='A'
goal='G'
def A_start(start,goal):
    visited=set()
    q=PriorityQueue()
    q.put((h[start],0,start))
    path=[]
    while not q.empty():
        f,g,node=q.get()
        if node==goal:
            return g,path
        if node in visited:
            continue
        visited.add(node)
        path.append(node)
        for n,n_cost in graph[node]:
            if n not in visited:
                g_new=n_cost+g
                f_new=g_new+h[n]
                q.put((f_new,g_new,n))
    return float('inf'),None
result=A_start(start,goal)
print(f"cost: {result[0]}")
print(f"path:{result[1]}")
        
```

```python
from queue import PriorityQueue

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('D', 5), ('E', 12)],
    'C': [('F', 7)],
    'D': [('G', 3)],
    'E': [('G', 2)],
    'F': [('G', 5)],
    'G': []
}

# Heuristic values (h)
h = {
    'A': 10, 'B': 8, 'C': 5, 'D': 7,
    'E': 3, 'F': 3, 'G': 0
}

def a_star(start, goal):
    pq = PriorityQueue()
    pq.put((h[start], 0, start, [start]))  # (f = g + h, g, node, path)
    visited = set()

    while not pq.empty():
        f, g, node, path = pq.get()

        if node == goal:
            return path, g

        if node in visited:
            continue
        visited.add(node)

        for neighbor, cost in graph[node]:
            if neighbor not in visited:
                g_new = g + cost
                f_new = g_new + h[neighbor]
                pq.put((f_new, g_new, neighbor, path + [neighbor]))

    return None, float('inf')

path, cost = a_star('A', 'G')
print("Path:", path)
print("Cost:", cost)

so what about this code?
```

```python
## min_max algorithm
class Node:
    def __init__(self,value):
        self.value=value
        self.children=[]
leaf1=Node(12)
leaf2=Node(34)
leaf3=Node(15)
leaf4=Node(10)
leaf5=Node(9)
leaf6=Node(11)
node1=Node(1)
node1.children=[leaf1,leaf2]
node2=Node(2)
node2.children=[leaf3,leaf4]
node3=Node(3)
node3.children=[leaf5,leaf6]
root=Node(0)
root.children=[node1,node2,node3]

def minmax(node,ismax,path):
    if not node.children:
        return node.value,path
    if ismax:
        best_val=float("-inf")
        best_path=[]
        for c in node.children:
            new_val,new_path=minmax(c,not ismax,path+[node])
            if new_val>best_val:
                best_val=new_val
                best_path=new_path
        return best_val,best_path
    else:
        best_val=float("inf")
        best_path=[]
        for c in node.children:
            new_val,new_path=minmax(c,not ismax,path+[node])
            if new_val<best_val:
                best_val=new_val
                best_path=new_path
        return best_val,best_path
value, path = minmax(root, True, [])
print("Optimal value:", value)
print("Path (node values):", [n.value for n in path])
        
```

```python
## alpha beta pruning
## min_max algorithm
class Node:
    def __init__(self,value):
        self.value=value
        self.children=[]
leaf1=Node(12)
leaf2=Node(34)
leaf3=Node(15)
leaf4=Node(10)
leaf5=Node(9)
leaf6=Node(11)
node1=Node(1)
node1.children=[leaf1,leaf2]
node2=Node(2)
node2.children=[leaf3,leaf4]
node3=Node(3)
node3.children=[leaf5,leaf6]
root=Node(0)
root.children=[node1,node2,node3]

def minmax(node,ismax,path,alpha,beta):
    if not node.children:
        return node.value,path
    if ismax:
        best_val=float("-inf")
        best_path=[]
        for c in node.children:
            new_val,new_path=minmax(c,not ismax,path+[node],alpha,beta)
            if new_val>best_val:
                best_val=new_val
                best_path=new_path
            alpha=max(alpha,best_val)
            if alpha>=beta:
                break
        return best_val,best_path
    else:
        best_val=float("inf")
        best_path=[]
        for c in node.children:
            new_val,new_path=minmax(c,not ismax,path+[node],alpha,beta)
            if new_val<best_val:
                best_val=new_val
                best_path=new_path
            beta=min(beta,best_val)
            if alpha>=beta:
                break
                
        return best_val,best_path
value, path = minmax(root, True, [],float('-inf'),float('inf'))
print("Optimal value:", value)
print("Path (node values):", [n.value for n in path])
        
```
