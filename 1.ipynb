```python
## best first search
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv('ToyotaCorolla.csv')
```

```python
data.head(2)
```

```python
## scatter plot
x=data["Price"]
y=data["Guarantee_Period"]
plt.scatter(x,y,cmap="jet")
plt.xlabel("Price")
plt.ylabel("Guarantee_Period")
plt.title("Scatter plot")
```

```python
## box plot
import seaborn as sns
```

```python
plt.boxplot([data['Price'],data['ABS']])

plt.xticks([1,2],["price","ABS"])
```

```python
## heatmap
sns.heatmap(data[['Price','Weight','Cylinders']].corr(),cmap="jet",annot=True)
```

```python
## coontor plot
x=data["KM"]
y=data['Weight']
z=data['Price']
```

```python
plt.tricontour(x,y,z,levels=20,cmap="jet")
plt.colorbar(label="Price")
```

```python
## 3d surface plot
```

```python
axes=plt.axes(projection="3d")
axes.plot(x,y,z)
axes.plot_trisurf(x,y,z,cmap="jet")
```
